```python
from osgeo import gdal,osr
import os 
```

```python
data_path=r'D:\contentone\yolov5\LargeImage'
all_big_images=os.listdir(data_path)
for data in all_big_images:
    dataset=gdal.Open(f'{data_path}\\{data}')
    prj = dataset.GetProjection()
    geot = dataset.GetGeoTransform()
    x_min = geot[0]
    y_max = geot[3]
    res = geot[1]
    srs = osr.SpatialReference(wkt = prj)
    x_length = res*dataset.RasterXSize
    y_length = res*dataset.RasterYSize
    if(srs.GetAttrValue('projcs')):
        x_length1 = x_length
        y_length1 = y_length
    else:
        x_length1 = x_length*111000
        y_length1 = y_length*111000
    if(x_length1<600):
        x_parts = 1
    else:
        x_parts = int(round(x_length1/600))
    if(y_length1<600):
        y_parts = 1
    else:
        y_parts = int(round(y_length1/600))

    x_size = x_length/x_parts
    y_size = y_length/y_parts
    x_stops = [x_min+ i*x_size for i in range(x_parts+1)]
    y_stops = [y_max - i*y_size for i in range(y_parts+1)]
    dataname=data.split('.')[0]
    for x in range(x_parts):
        for y in range(y_parts):
            xmin = x_stops[x]
            xmax = x_stops[x+1]
            ymax = y_stops[y]
            ymin = y_stops[y+1]
            gdal.Warp(f'{data_path}\\{dataname}_part{x}{y}'+'.tif',dataset,outputBounds = (xmin,ymin,xmax,ymax))
    del dataset
```
